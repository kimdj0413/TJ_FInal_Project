```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import re
```

```python

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'ko-KR,ko;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive'
}
people_link = []
query = 0

while True:
    patient = 0
    query += 1
    url = f'https://kin.naver.com/people/partner/index.naver?dirId=4&page={query}'

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    links = soup.find_all('a', class_='spot_partner__name')

    # 각 링크의 href 값을 출력하기
    for link in links:
        if patient < 5:
            patient += 1
        else:
            address = 'https://kin.naver.com/'+link['href']
            people_link.append(address)

    if query == 10:
        break
print(people_link)
print(len(people_link))
```

```python
people_more_link = []
for url in people_link:
    url = url
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    link = soup.find('span', class_='more')
    a_tag = link.find('a')
    href_value = 'https://kin.naver.com/' + a_tag['href']
    people_more_link.append(href_value)
        
print(people_more_link)
print(len(people_more_link))
```

```python
# test_links = people_more_link[1]
# test_links = test_links + '&page='

question_list = []

for test_link in people_more_link:
    for i in range(1,100):
        query = i
        url = f'{test_link}&page={query}'
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        links = soup.find_all('td', class_='title')
        for link in links:
            a_tag = link.find('a')
            href_value = 'https://kin.naver.com/' + a_tag['href']
            question_list.append(href_value)
print(question_list)
print(len(question_list))
```
