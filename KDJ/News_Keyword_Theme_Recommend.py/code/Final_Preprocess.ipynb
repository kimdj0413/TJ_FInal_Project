```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import re
from collections import defaultdict
import pickle
from konlpy.tag import Okt
from sentence_transformers import SentenceTransformer
import os
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'ko-KR,ko;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive'
}

url = 'https://www.judal.co.kr/?view=themeList'
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
elements = soup.find_all(class_='btn p-0 m-0 text-left')

print(elements)
print(len(elements))
```

```python
theme_name = []

for element in elements:
  theme_name.append(element.get_text().replace('\n',''))

hrefs = [a['href'] for a in soup.find_all('a', class_='btn p-0 m-0 text-left', href=True)]

print(theme_name)
print(hrefs)
print(len(hrefs))
```

```python
theme_dict = defaultdict(list)

for i in range(0, len(theme_name)):
  href = hrefs[i]
  href = href.replace('amp;','')
  stock_response = requests.get(href, headers=headers)
  stock_soup = BeautifulSoup(stock_response.text, 'html.parser')
  stock_elements = stock_soup.find_all(class_='btn p-0 m-0 text-left')
  for stock in stock_elements:
    stock = stock.find('b').get_text()
    theme_dict[theme_name[i]].append(stock)
    
print(theme_dict)
```

```python
with open('D:/Topic_Modeling/New/theme_dict.pkl', 'wb') as f:
    pickle.dump(theme_dict, f)
    # print('저장완료')
formatted_keys = ', '.join(theme_dict.keys()).replace(', ', '\n')

# 결과 출력
print(formatted_keys)
```

```python
theme_dict = {key.replace('/', '-') : value for key, value in theme_dict.items()}

with open('D:/Topic_Modeling/New/theme_dict.pkl', 'wb') as f:
    pickle.dump(theme_dict, f)

with open('D:/Topic_Modeling/New/theme_dict.pkl', 'rb') as f:
    theme_dict = pickle.load(f)

formatted_keys = ', '.join(theme_dict.keys()).replace(', ', '\n')

# 결과 출력
print(formatted_keys)
```
